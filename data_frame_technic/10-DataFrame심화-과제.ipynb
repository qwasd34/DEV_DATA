## 수익을 다음으로 분류
- 60000이상 고임금
- 30000이상~60000미만 평균임금
- 30000미만 저임금
- 열 이름은 'Class_Income'

```python
def money(n):
    if n>=60000:
        return '고임금'
    elif n>=30000 and n<60000:
        return '평균'
    elif n<30000:
        return '저임금'
    else:
        return '---'
```

```python
money(100), money(40000), money(10000), money(66666)
```

```python
cus['Class_Income']=cus['Income'].apply(money)
cus
```

```python
data=cus.pop('Class_Income')
data
```

```python
cus.insert(3, 'Class_Income', data)
cus
```

## Alone, Absurd, YOLO를 Single로 변경
1. reset_index()로 되돌리고
2. replace으로 값을 바꾼다음에
3. 다시 set_index()로 설정해보자.
4. 실제로 잘 변경되었는지 get_level_values(0)후 value_counts()로 데이터 경우의 수를 확인해보자.

멀티 인덱스일때 set_level로 할수 있지만 너무 비효율적이다. 
인덱스를 열로 되돌린 다음에 진행!! 

```python
cus.reset_index(inplace=True)
```

```python
cus.head()
```

```python
#2. replace으로 값을 바꾼다음에
cus['Marital_Status']=cus['Marital_Status'].replace(['Alone', 'Absurd', 'YOLO'], 'Single')
cus['Marital_Status'].value_counts()
```

```python
# 3.다시 set_index()로 설정해보자.
cus=cus.set_index(['Marital_Status','Education'])
cus
```

```python
#4. 실제로 잘 변경되었는지 get_level_values(0)후 value_counts()로 데이터 경우의 수를 확인해보자.
cus.index.get_level_values(0)
```

```python
cus.index.get_level_values(0).value_counts()
```
