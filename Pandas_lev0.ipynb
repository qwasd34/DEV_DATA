````python
# 패스 능력치 별 평가
def check_skill_pass(x):
  if x>=90:
    return '우수'
  if x >= 70:
    return '보통'
  return '미흡'
````

````python
# 함수 적용하여 새로운 열 추가하기
df_power["패스평가"]=df_power["패스"].apply(check_skill_pass)
df_power.head()
````

```
  df_power['패스평가'] = df_power['패스'].apply(check_skill_pass)
```
에서 SettingWithCopyWarning: 은
- 이 경고는 Pandas에서 데이터프레임의 일부를 선택한 후 값을 변경할 때 발생하는 경고 메시지
- 값의 계산과 수정이 동시에 일어날 때의 경고
- 때에 따라 예상과 다른 동작이나 메모리 부족 상태가 발생할 수 있다.
- [(참고)Pandas의 DataFrame 사용 중 SettingWithCopyWarning 발생할 때](https://emilkwak.github.io/pandas-dataframe-settingwithcopywarning)


````python
# 파이썬에서 for과 원본데이터를 조작할 때 의도치 않은 상황 (파이썬 튜터)
numbers=[0,1,2,3,4,5,6,7]
for n in numbers:
  print(n)
  numbers.remove(n)
print(numbers)
````

````python
# 파이썬에서 for을 이용한 유사한 상황 - .copy()나 [:]로 해결
numbers=[0,1,2,3,4,5,6,7]
for n in numbers.copy():
  print(n)
  numbers.remove(n)
print(numbers)
````

## 피봇테이블 - pivot_table

````python
# 본래 df를 확인
df.head()
````

````python
# '신장'을 index로 지정한 pivot_df 생성하기
# aggfunc : 집계함수(aggregate function)
# aggfunc=np.sum, aggfunc=sum, aggfunc='sum'도 모두 되긴 한다.
pivot_df=pd.pivot_table(df, index='신장',values=['3점슛','골밑슛'], aggfunc='mean')
pivot_df.round(2)
````

## 얕은 복사(Shallow Copy) vs 깊은 복사(Deep Copy)

````python
# 피봇으로 만든 데이터프레임 조회
pivot_df.head()
````

````python
# 1. 얕은 복사 2. 깊은 복사
pivot_df_shallow=pivot_df
pivot_df_copy=pivot_df.copy()
````

````python
# 얕은 복사는 변경이 일어나면 원본도 변경됨
del pivot_df_shallow['골밑슛']
````

````python
# 얕은 복사는 변경이 일어나면 원본도 변경됨 - 메모리 주소 공유, 사실상 변수명만 다른 것
pivot_df_shallow.head()
pivot_df.head()
````

````python
#얕은복사!! 주소만 공유
a=[0,3,6,9]
b=a
del b[0]
del b[-1]
print(a)
print(b)
````

````python
#깊은복사
a=[0,3,6,9]
b=a.copy()
del b[0]
del b[-1]
print(a)
print(b)
````

````python
# 깊은 복사는 별개의 df기에 원본에 영향을 주지 않음
pivot_df_copy
````

````python
# deep copy의 변경을 시켜보면 pivot_df_copy는 변했지만
del pivot_df_copy['3점슛']
````

````python
# pivot_df는 변하지 않는다.
pivot_df_copy
pivot_df
````
