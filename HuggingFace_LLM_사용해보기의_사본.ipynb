## Use the HuggingFaceHub

```python
from langchain import HuggingFaceHub
```

```python
llm=HuggingFaceHub(
        repo_id="google/flan-t5-large",
        model_kwargs={
            "temperature":90,  # 0-100. 0에 가까우면 더 확률이 높은 답을 하고 100에 가까운 랜덤해짐
            "max_length":64
        }
    )
```

```python
question = "What is the capital of France?"

print(llm.invoke(question))
```

```python
question = "What area is best for growing wine in France?"

print(llm.invoke(question))
```

```python
question = "What is the capital of England?"

print(llm.invoke(question))
```

```python
question = "What is the capital of Korea?"

print(llm.invoke(question))
```

## T5-Flan - Encoder-Decoder

```python
from transformers import AutoTokenizer, pipeline, AutoModelForSeq2SeqLM

model_id = 'google/flan-t5-large'# go for a smaller model if you dont have the VRAM
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForSeq2SeqLM.from_pretrained(model_id) # , load_in_8bit=True)

local_llm = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=100
)
```

```python
print(local_llm('What is the capital of Korea? '))
```

```python
print(local_llm("What is the population of Korea?"))
```

### GPT2-medium - Decoder Only Model


```python
from transformers import pipeline
```

```python
model_id = "gpt2-medium"

local_llm = pipeline(
    "text-generation",
    model = model_id,
    max_length=100
)
```

```python
question = "kpop is"
answers = local_llm(question, max_length=30, num_return_sequences=5)
```

```python
for a in answers:
    print(a)
```
