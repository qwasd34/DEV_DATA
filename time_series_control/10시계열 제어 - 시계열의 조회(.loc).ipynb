## 조회
- .iloc인 사용방법은 동일
- .loc도 사용방법은 동일한데, label을 어떻게 입력해야할까?
- 기간 탐색의 장점을 살려보자.

```python
# 년월일 .loc['label'] 풀매칭이 아니여도됨 
stocks.loc[['2015-05-27']]	
```

```python
# .loc[['label1', 'label2']] 
stocks.loc[['2015-05-27', '2015-6']]	
```

```python
# 슬라이싱 1
stocks.loc[:'2015-06-1']
```

```python
# 슬라이싱 2
stocks.loc['2015-05-27': '2015-6']
```

```python
# 년월만 탐색
stocks.loc['2015-05']
```

```python
# 년월 슬라이싱
stocks.loc['2015.06':'2015.07']
```

```python
# 년 조회
stocks.loc['2015']
```
