| 컬럼명 | 설명  |
| :-- | :---- |
| symbol| 애플 주식  |
| close | 종가. 종료가  |

| low | 그 날의 최저가  |
| open  | 그 날의 시가. 시작가  |
| volume| 거래량 |
| adjClose| 다른 주식 속성/행동과 관련하여 조정된 종가  |

| adjVolume  | 다른 주식 속성/행동과 관련하여 조정된 거래량 |
| divCash | 현금 배당  |
| splitFactor  | 주식 분할  |


```python
cols = ['date', 'open', 'high', 'low', 'close']
```

```python
stocks=pd.read_csv('AAPL.csv' ,usecols=cols, index_col='date')
```

```python
stocks
```

## index > DatetimeIndex : to_datetime()

```python
# 현재 index의 dtype? 문자열
stocks.index
```

```python
# 방법 1. index의 속성 변경 to_datetime()
pd.to_datetime(stocks.index)
```

```python
stocks.index=pd.to_datetime(stocks.index)
stocks
```

## read_csv(parse_dates=)

```python
# 방법 2.
cols = ['date', 'open', 'high', 'low', 'close']
```

```python
stocks=pd.read_csv('AAPL.csv' ,usecols=cols, index_col='date', parse_dates=['date'])
```

```python
stocks
```

```python
stocks.index
```

## 복습 : 요일 열 추가

```python
days=stocks.index.day_name(locale='ko_KR.utf8')
days
```

```python
stocks.insert(0, '요일', days)
```
