## 이름(name) 변경

```python
# 참조
import pandas as pd
```

```python
# 데이터프레임 생성
cols = ['name', 'survived', 'pclass', 'fare', 'sex', 'age']
tt = pd.read_excel('titanic3.xls', usecols=cols, index_col='name')
tt
```

## 열(column) 이름 변경

```python
# 열 목록(columns index) 조회
tt.columns
```

```python
# 특정 열 이름 접근
tt.columns[0]
```

```python
# Try : 수정 = 접근 후 할당 
tt.columns[0]='class'
```

```python
# 방법 1 : rename을 이용하기
tt.rename(columns={'pclass' : 'class'}, inplace=True)
```

```python
tt.head()
```

```python
# 방법 2 : 같은 길이의 리스트를 준비해서 교체하기
tt.columns
```

```python
# 길이가 5이니 5개 요소의 리스트 준비하기, 비용(fare->cost)로 변경
tt.columns=['class', 'survived', 'sex', 'age', 'cost']
```

```python
tt.columns
```

```python
tt.head()
```

## rename() 좀 더 살펴보기 - 행과 열의 인덱스 이름 변경
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename.html
- https://pandas.pydata.org/docs/reference/api/pandas.Series.rename.html
- https://pandas.pydata.org/docs/reference/api/pandas.Index.rename.html

```python
# 1. 2개의 열을 변경하려면? columns 대신 axis활용
tt.rename({'class': 'pclass', 'cost': 'fare'} ,axis=1)
```

```python
# axis인자를 빼고 한다면
tt.rename(columns={'class': 'pclass', 'cost': 'fare'} )
```

```python
# 2. 인덱스에서 'Allen, Miss. Elisabeth Walton' 이름 변경
tt.rename({'Allen, Miss. Elisabeth Walton': 'Allen'}, axis=0)
```

```python
# axis인자를 빼고 한다면
tt.rename(index={'Allen': 'Allen, Miss. Elisabeth Walton'})
```

```python
# 3. 행렬 인덱스를 설정
tt.index.name= 'passenger'
tt.columns.name='Info'
```
