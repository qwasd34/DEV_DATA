## Pandas참조 후 titanic 데이터를 가져오기
- name열이 index로 설정

```python
import pandas as pd
tt=pd.read_excel('titanic3.xls', index_col='name')
tt
```

## 데이터프레임의 columns만 조회

```python
tt.columns
```

## .drop()을 이용해서 2개의 열을 삭제
- home.dest
- embarked
- sibsp
- 원본에 반영

```python
tt.drop(columns=['home.dest', 'embarked', 'sibsp'], inplace=True)
tt.head()
```

## del을 이용한 열 삭제
- parch

```python
del tt['parch']
tt.head()
```

## 사망자 명단 조회하기

```python
dead_mask=tt['survived']==0
tt[dead_mask]
```

## 시신 식별번호(body)가 없는 데이터 조회하기

```python
tt.info()
```

```python
tt['body'].isna().sum()
```

```python
unkown_body_mask=tt['body'].isna()
tt[unkown_body_mask]
```

살아있는 사람은 시신 식별번호가 없는 것은 당연

## 사망자이나 시신식별번호가 없는 명단 찾기

```python
tt[dead_mask & unkown_body_mask]
```

사망자가 809명인데, 식별번호가 없는 명단이 688명이나 된다.

## 사망자이나 시신식별번호가 없는 명단으로 새로운 df 만들기
- `notfound_list`

```python
notfound_list=tt[dead_mask & unkown_body_mask]
notfound_list
```

## notfound_list의 승객 등급별 개수 확인

```python
notfound_list['pclass'].value_counts()
#잘못된 분석! 
```

3등급(하류층)이 많으나, 이건 당연할 수 있다. 비율을 구해보자.

## 전체 titanic 전체 승객 등급별 개수 확인

```python
tt['pclass'].value_counts()
```

## 승객 등급별 실종 비율 확인

```python
(notfound_list['pclass'].value_counts() / tt['pclass'].value_counts())*100
```

## 1. 사망자의 하류층 티겟 가격 분포와 2. 생존자의 하류층 티켓 가격 분포를 확인해보자.

```python
survived_mask=tt['survived']==1
pclass_3_mask=tt['pclass']==3
```

```python
dead_pclass3_mask=tt[dead_mask & pclass_3_mask]
#dead_pclass3_mask['fare'].value_counts(normalize=True)
```

```python
survived_pclass3_mask=tt[survived_mask & pclass_3_mask]
#survived_pclass3_mask.value_counts(subset='fare',normalize=True)
```

```python
dead_pclass3_mask.describe()
```

```python
survived_pclass3_mask.describe()
```

3등급에서 티켓가격의 차이가 생존유무와 관련성이 있어보이지 않는다.
