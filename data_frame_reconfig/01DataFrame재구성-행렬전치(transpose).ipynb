## .transpose()
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transpose.html
- 전치행렬 : 행과 열을 교환하여 얻는 행렬. 주대각선을 축으로 반사 대칭을 한다.
- 행과 열을 반전시키자.

![Transpose](Matrix_transpose.gif "transpose")

```python
# 전치를 해보면.. 행열 바뀜
student.transpose()
```

```python
# 줄여서 .T
student.T
company.T
```

위의 전치는 유의미하지 않다!
데이터의 증가는 행의 방향이 옳다고 생각될 경우에 유의미할 것

```python
league.T
```

## 주의사항 : 원본 != 전치*2
- 값은 같아보여도 dtype도 같을까?

```python
# 원본
company.info()
company.head()
```

```python
# 전치 1회
t1_com=company.T
```

```python
t1_com.info()
t1_com.head()
```

```python
# 전치 2회 dtype 이 다 문자열로 변화되어있어
t2_com=t1_com.T
```

```python
t2_com.info()
t2_com.head()
```

전치 2번을 했더니 모든 열의 dtype이 object로 변경되었다!
