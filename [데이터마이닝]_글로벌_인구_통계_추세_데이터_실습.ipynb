## ìë ê²½ë ì ë³´ ì¶ê°

- ì§ëì ì ë³´ë¥¼ ë½ìì¤ë ¤ë©´ ìë ê²½ë íì   dataset ãì ì ë³´ê° ìì´ì Geopy - ëë¼ì´ë¦ë³ ìë ê²½ë ê°ì ¸ì¬ììì

```python
# ###########################################################
# ë°©ë² 1) APIë¥¼ íµí´ í¸ì¶íë ê²½ì° ìê°ì´ ì¤ëê±¸ë¦¬ê³ , ì íì´ ê±¸ë¦´ ì ìì! #
# ###########################################################

from geopy.geocoders import Nominatim
import time

# Geopyì Nominatim ì¬ì©ì ìí Geolocator ê°ì²´ ìì±
geolocator = Nominatim(user_agent="geoapiExercises")

# êµ­ê°ë³ ìëì ê²½ëë¥¼ ì ì¥í  ëìëë¦¬ ìì±
location_coordinates = {}

# ì¤ë³µë ìì¹ ìì²­ì ë°©ì§íê¸° ìí´ ì´ë¯¸ ì¡°íë ìì¹ë¥¼ ì ì¥í  ì§í©
queried_locations = set()

# ê° êµ­ê°ì ëí´ ìëì ê²½ë ì¡°í
for location in all_countries:
    # ì´ë¯¸ ì¡°íë ìì¹ë ê±´ëë
    if location in queried_locations:
        continue

    try:
        # ìì¹ì ëí ìë, ê²½ë ì ë³´ ì»ê¸°
        loc = geolocator.geocode(location)
        if loc:
            location_coordinates[location] = (loc.latitude, loc.longitude)
            queried_locations.add(location)
    except Exception as e:
        print(f"Error occurred for location: {location}, error: {e}")

    # ìì²­ ì¬ì´ì ì¼ì  ìê° ì§ì°ì ëì´ ìë¹ì¤ ì íì í¼í¨
    time.sleep(1)

location_coordinates
```

```python
# ########################################
# ë°©ë² 2) ì´ë¯¸ ë§ë¤ì´ë ìë & ê²½ë ë°ì´í° ë¶ë¬ì¤ê¸° #
# ########################################

# ë¤ì´ë¡ë ë°ì ë°ì´í° ìë¡ë íì!
import json

with open('location_coordinates.json', 'r') as f:
    location_coordinates = json.load(f)
```

```python
location_coordinates
```

```python
# ë°ì´í° íë ìì ìë ë° ê²½ë ì ë³´ ì¶ê°
country_data['latitude'] = country_data['Location'].apply(lambda x: location_coordinates[x][0]
                                                          if x in location_coordinates else None)
country_data['longitude'] = country_data['Location'].apply(lambda x: location_coordinates[x][1]
                                                           if x in location_coordinates else None)

country_data.head()
```

## ì¬ë¼ì´ë íì©

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# ì°ë ì íì ìí ì¬ë¼ì´ë ìì ¯ ìì±
year_slider = widgets.IntSlider(
    value=2020,
    min=1950,
    max=2022,
    step=1, #ì¼ë§í¼ ìì§ì¼ì§
    description='Year:',
    continuous_update=False # ê° ë³íê° ìë ê·¸ ìê°ì ë³íê° ìë°ì´í¸ ëì§ ìëë¡ í¨. ë³íê° ë©ì¶°ì¼ ê°ì ìë°ì´í¸
)

# ì§ë ì¶ë ¥ì ìí Output ìì ¯
map_output = widgets.Output()

data_clean = country_data.dropna(subset=['latitude', 'longitude'])

def update_map(year):
    with map_output:
        # ì¶ë ¥ ìì­ì í´ë¦¬ì´íê³  ì ì§ëë¥¼ íì
        clear_output(wait=True)
        # ì§ë ìì±
        m = folium.Map(location=[20, 0], zoom_start=2)
        # í´ë¹ ì°ëì ë°ì´í° íí°ë§
        data_year = data_clean[data_clean['Time'] == year]

        # ì§ëì ë°ì´í° ì¶ê°
        for idx, row in data_year.iterrows():
            radius = np.sqrt(row['TPopulation1July']) * 1000
            folium.Circle(
                location=[row['latitude'], row['longitude']],
                radius=radius,
                color='blue',
                fill=True,
                fill_color='blue',
                popup=f"{row['Location']}: {row['TPopulation1July']}"
            ).add_to(m)

        # ì§ë íì
        display(m)

def on_year_change(change):
    update_map(change['new'])

year_slider.observe(on_year_change, names='value')

# ì´ê¸° ì§ë ë° ì¬ë¼ì´ë ìì ¯ íì
display(year_slider)
update_map(2020)
display(map_output)
```

```python
import seaborn as sns #matplotlib ê¸°ë°ì¼ë¡ í¨
import ipywidgets as widgets
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

year_slider = widgets.IntSlider(
    value=2020,
    min=1950,
    max=2022,
    step=1,
    description='Year:',
    continuous_update=False
)

data_clean = country_data.dropna(subset=['latitude', 'longitude'])

# ë° ì°¨í¸ ì¶ë ¥ì ìí Output ìì ¯
output = widgets.Output()

# ë° ì°¨í¸ë¥¼ ê·¸ë¦¬ë í¨ì
def plot_top10_population(year):
    with output:
        clear_output(wait=True)
        # í´ë¹ ì°ë ë°ì´í° íí°ë§
        data_year = data_clean[data_clean['Time'] == year]

        # ìì 10ê° êµ­ê°ì ì¸êµ¬ ë°ì´í° ì¶ì¶
        top10_data = data_year.nlargest(10, 'TPopulation1July')

        # ë° ì°¨í¸ ê·¸ë¦¬ê¸°
        plt.figure(figsize=(10, 6))
        sns.barplot(x='TPopulation1July', y='Location', data=top10_data) #í¸íì´ë¨
        plt.title(f'Top 10 Countries by Population in {year}')
        plt.xlabel('Population (in thousands)')
        plt.ylabel('Country')
        plt.show()

# ì¬ë¼ì´ëì ê°ì´ ë³ê²½ë  ëë§ë¤ ë° ì°¨í¸ ìë°ì´í¸
def on_year_change(change):
    plot_top10_population(change['new'])

year_slider.observe(on_year_change, names='value')

# ì´ê¸° ë° ì°¨í¸ ë° ì¬ë¼ì´ë ìì ¯ íì
display(year_slider)
plot_top10_population(2020)
display(output)
```

## ê³ ì  ììê° ì§ì  Bar Plot

```python
import seaborn as sns
import ipywidgets as widgets
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

year_slider = widgets.IntSlider(
    value=2020,
    min=1950,
    max=2022,
    step=1,
    description='Year:',
    continuous_update=False
)

data_clean = country_data.dropna(subset=['latitude', 'longitude'])

# ì ì²´ ë°ì´í°ììì ê³ ì í ëë¼ë¤ ì¶ì¶
unique_countries = data_clean['Location'].unique()

# ê³ ì í ëë¼ë¤ì ëí ìì ë§¤í ìì±
colors = sns.color_palette("tab20", len(unique_countries))
country_colors = dict(zip(unique_countries, colors))

output = widgets.Output()

def plot_top10_population(year):
    with output:
        clear_output(wait=True)
        data_year = data_clean[data_clean['Time'] == year]
        top10_data = data_year.nlargest(10, 'TPopulation1July')

        # ìì 10ê° êµ­ê°ì ëí ìì í ë¹
        top10_colors = {country: country_colors[country] for country in top10_data['Location']}

        # ë° ì°¨í¸ ê·¸ë¦¬ê¸°
        plt.figure(figsize=(10, 6))
        sns.barplot(x='TPopulation1July', y='Location', data=top10_data, hue='Location', legend=False,
                    palette=[top10_colors[country] for country in top10_data['Location']])
        plt.title(f'Top 10 Countries by Population in {year}')
        plt.xlabel('Population (in thousands)')
        plt.ylabel('Country')
        plt.show()


def on_year_change(change):
    plot_top10_population(change['new'])

year_slider.observe(on_year_change, names='value')

display(year_slider)
plot_top10_population(2020)
display(output)
```
