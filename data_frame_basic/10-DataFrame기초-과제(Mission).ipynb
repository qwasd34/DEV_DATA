## Pandas참조합니다. (별칭 사용)

```python
import pandas as pd
```

## imdb 영화 순위 Top1000정보가 들어있는 imdb_top_1000.csv 로 DataFrame만들기

```python
mdf=pd.read_csv('imdb_top_1000.csv')
mdf.head()
```

## 필요한 열만 가져오기
- 아래 순서로 열이 정렬되야합니다.

'Genre', 'Series_Title', 'Released_Year', 'IMDB_Rating', 'Meta_score', 'Director', 'No_of_Votes', 'Gross'

```python
cols=['Genre', 'Series_Title', 'Released_Year', 'IMDB_Rating', 'Meta_score', 'Director', 'No_of_Votes', 'Gross']
```

```python
mdf=mdf[cols]
mdf.head()
```

## 데이터프레임 정보를 확인합시다.
- 각 dtype과 null 여부를 확인합시다.

```python
mdf.info()
```

## Top1000에 들어간 영화에서 가장 많은 장르 5개를 확인해봅시다.

```python
mdf['Genre'].value_counts().nlargest(5)
```

```python
#mdf.nlargest(n=5, columns='Genre', keep='first')
```

## 수익이 측정이 안된 영화는 제외하여 저장합시다.
- Meta_score는 NA여도 삭제하지 않습니다.

```python
mdf=mdf.dropna(subset='Gross')
mdf.info()
```

## 메타스코어 평균을 확인합니다.

```python
mdf['Meta_score'].mean().round(1)
```

## 가장 수익을 많이 낸 영화 Top1을 찾아봅시다.
- 힌트 : 현재 수익을 나타내는 Gross의 dtype을 주의하자.
- 힌트 : 문자열은 크기 정렬이 안된다.

```python
# 1. 쉼표제거
mdf['Gross']=mdf['Gross'].str.replace(",", "")
mdf.head()
```

```python
# 타입변경 o->int
mdf['Gross']=pd.to_numeric(mdf['Gross'])
mdf.info()
mdf.head()
```

```python
mdf.nlargest(n=1, columns=['Gross'])
```

## 가장 수익을 많이 낸 영화 Top10을 확인해봅시다.

```python
mdf.nlargest(n=10, columns=['Gross'])
```

## 가장 수익을 많이 낸 영화 Top100의 장르 개수를 확인해봅시다.

```python
mdf.nlargest(n=100, columns=['Gross']).value_counts('Genre')
```

## 가장 수익이 적은 영화 Top1을 찾아봅시다.

```python
mdf.nsmallest(n=1, columns=['Gross'])
```

## 가장 수익이 적은 영화를 Top10을 확인해봅시다.

```python
mdf.nsmallest(n=10, columns='Gross')
```
