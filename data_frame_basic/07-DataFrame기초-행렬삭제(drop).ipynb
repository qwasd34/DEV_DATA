## 데이터 준비
- https://www.kaggle.com/datasets/harshitshankhdhar/imdb-dataset-of-top-1000-movies-and-tv-shows/

```python
import pandas as pd
```

```python
# 가져올 열 지정하기
mdf = pd.read_csv('imdb_top_1000.csv')
mdf
```

## .drop() - 1. 열 삭제(columns=)
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html
- 불필요해보이는 Poster_link와 Overview를 삭제하자.

```python
mdf.drop(columns=['Poster_Link','Overview'], inplace=True)
mdf
```

## .drop() - 2. 행 삭제(by label)

```python
# index label이 1인 것을 삭제
mdf.drop(labels=1, inplace=True)
mdf.head()
```

## .drop() - 3. 행 삭제(by labels, list-like)

```python
# 삭제할 index label을 리스트로 전달
# movie_df.drop(labels=[2, 3, 4], inplace=True)
mdf.drop([5, 7, 9], inplace=True)
mdf
```

## .drop() - 4. 열 삭제(axis=1 필요)

**다시보는 axis설명**
- axis = 0 : index 방향
- axis = 1 : columns 방향

```python
mdf.drop(['Star1','Star2','Star3','Star4'], axis=1, inplace=True)
mdf
```

## errors= : .drop() 에러 제어

**errors : {‘ignore’, ‘raise’}, default ‘raise**  
If ‘ignore’, suppress error and only existing labels are dropped.

```python
# 이미 삭제한 데이터라 에러 발생
mdf.drop([4,5,6] ,inplace=True)
mdf
```

```python
# 이미 삭제한 데이터지만 에러가 발생안 함
mdf.drop([4,5,6] ,inplace=True, errors='ignore')
mdf
```

## del을 이용한 열 삭제

```python
# Runtime을 삭제를 한다고 할 때
del mdf['Runtime']
mdf.head()
```

```python
# 이미 삭제된, 없는 열을 삭제 시도하면? -> 에러 발생
del mdf['Runtime']
```

1. 해당 데이터 주소로 찾아가서
2. del로 그 할당을 삭제

[생각하기] `del`은 리스트 전용, 딕셔너리 전용 삭제 키워드가 아니다.
