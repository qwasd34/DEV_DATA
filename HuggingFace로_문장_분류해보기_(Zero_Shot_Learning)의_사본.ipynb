텍스트 감정 분류
- Zero shot 분류
- transformers 라이브러리 사용
  - 사용방식은 분류 레이블을 제공하면서 분류 대상이 되는 텍스트 제공
  

```python
from transformers import pipeline
```

text classification
https://huggingface.co/facebook/bart-large-mnli

```python
classifier = pipeline(model="facebook/bart-large-mnli")
```

```python
classifier("one day I will see the world",
    candidate_labels=['travel', 'cooking', 'dancing', 'exploration'],
    multi_label=False
)
```

```python
classifier("I have a problem with my iphone that needs to be resolved asap!",
    candidate_labels=["positive", "neutral", "negative"],
    multi_label=True
)
```

```python
classifier("In the afternoon, I decided to go for a walk in the park.",
           candidate_labels=['positive','neutral','negative'],
           multi_label=True
           )
```
